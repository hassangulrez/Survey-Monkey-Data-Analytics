Create Array for irrelevant Columns to drop

```python
columns_to_drop= ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1']
```

Columns Break down to use melt function for ID and Value Variables

```python
list(surveydata.columns)[:8]
```

```python
list(surveydata.columns)[8:]
```

```python
id_vars=list(surveydata.columns)[:8]
```

```python
value_vars=list(surveydata.columns)[8:]
```

```python
surveydata_melt=surveydata.melt(id_vars=id_vars, value_vars=value_vars)
```

```python
surveydata_melt
```

Rename Variable and Value Column set to our Desired Names

```python
surveydata_melt=surveydata.melt(id_vars=id_vars, value_vars=value_vars, var_name="Questions + Subquestion", value_name="Answers" )
```

```python
surveydata_melt
```

Import Questionare1 worksheet  which contain 2 top rows and transposing into 2 columns and concatenate finally into  contain "SUBQUESTIONS" and "Questions+Subquestion" columns

```python
Questions_import = pd.read_excel('C:/Users/hassa/OneDrive/Desktop/SurveyMonkey/questionare1.xlsx')
```

```python
Questions_import
```

Make a copy of  Questions_import

```python
Questions=Questions_import.copy()
```

```python
Questions
```

Drop columns 'RAW QUESTIONS','RAW SUBQUESTIONS','SUBQUESTIONS' as we are interested only Questions and Question + Subquestion columns

```python
ValidQuestions=Questions.drop(columns=['RAW QUESTIONS','RAW SUBQUESTIONS','SUBQUESTIONS'])
```

```python
ValidQuestions
```

DROP NA values if Question , Questions + Subquestion dataset has any?

```python
ValidQuestions=ValidQuestions.dropna()
```

```python
ValidQuestions
```

Apply merge pandas formula to combine both dataset using left join and  "Questions + Subquestion" is present in both dataset

```python
dataset_merged=pd.merge(left=surveydata_melt, right=ValidQuestions, how="left", left_on="Questions + Subquestion", right_on="Questions + Subquestion")
print("Original Data", len(surveydata_melt))
print("Merged Data", len(dataset_merged))
dataset_merged
```

```python
#dataset_merged=dataset_merged.rename(columns={"RAW QUESTIONS.1":"Questions"}, inplace=True)
```

Remove na values from Answers Columns and use group by function to find number of Questions Correspond to RespondentsID 

```python
respondents = dataset_merged[dataset_merged["Answers"].notna()]
respondents = respondents.groupby("QUESTIONS")["Respondent ID"].nunique().reset_index()
respondents.rename(columns={"Respondent ID":"Respondents","RAW QUESTIONS.1":"QUESTTIONS"}, inplace=True)
```

```python
respondents
```

Use Merge formula again to find correlation using left join i.e QUESTIONS

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondents, how="left", left_on="QUESTIONS", right_on="QUESTIONS")
print("Original Data", len(dataset_merged))
print("Merged Data", len(dataset_merged_two))
dataset_merged_two
```

```python
same_answer = dataset_merged # [dataset_merged["Answer"].notna()]
same_answer = same_answer.groupby(["Questions + Subquestion", "Answers"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID":"Same Answer"}, inplace=True)
same_answer
```

```python
dataset_merged_three = pd.merge(left=dataset_merged_two, right=same_answer, how="left", left_on=["Questions + Subquestion", "Answers"], right_on=["Questions + Subquestion", "Answers"])
dataset_merged_three["Same Answer"].fillna(0, inplace=True)
print("Original Data", len(dataset_merged_two))
print("Merged Data", len(dataset_merged_three))
dataset_merged_three
```

```python
output = dataset_merged_three.copy()
output.rename(columns={"Identify which division you work in. - Response":"Division Primary", "Identify which division you work in. - Other (please specify)":"Division Secondary", "Which of the following best describes your position level? - Response":"Position", "Which generation are you apart of? - Response":"Generation", "Please select the gender in which you identify. - Response":"Gender", "Which duration range best aligns with your tenure at your company? - Response":"Tenure", "Which of the following best describes your employment type? - Response":"Employment Type"}, inplace=True)
output
```
